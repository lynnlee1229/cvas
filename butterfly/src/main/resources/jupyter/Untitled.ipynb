```python
import geopandas as gpd
import pandas as pd
import json
import os
```

```python
def gpd_read_csv(path,wkt_col='geometry'):
    df = pd.read_csv(path)
    s = gpd.GeoSeries.from_wkt(df[wkt_col])
    gdf = gpd.GeoDataFrame(data=df, geometry=s)
    return gdf
```

```python
gdf1=gpd_read_csv('/Users/lynnlee/Data/diploma/testin/shenzhen_landuse.csv')
gdf2=gpd_read_csv('/Users/lynnlee/Data/diploma/testin/shenzhen_district.csv')
```

```python
gpd.sjoin(gdf1,gdf2,how="left",predicate="intersects")
```

```python
gdf2.reset_index().to_csv('../outs/shenzhen_district.csv',index=False)
```
